#Unemployment Analysis with Python
Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19,So analysing the unemployment rate can be a good data science project.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv file
df=pd.read_csv('/content/Unemployment in India.csv')
df.head()   #displaying few rows
```

```python
#data preprocessing
df.isnull().sum()
df.rename(columns={df.columns[3]:'UnemploymentRate',df.columns[4]:'Employed',df.columns[5]:'LabourParticipationRate'},inplace=True )
df
```

```python
#data visualization
states_unemployment= df.groupby('Region')['UnemploymentRate'].mean()  #calculating mean
plt.figure(figsize=(8,7))
states_unemployment.sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Average unemployment rate by states in India ')
plt.xlabel('Regions')
plt.ylabel('Unemployment rate')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Region'].value_counts()   #frequency of each region
```

```python
#creating heatmap
heatmap_data= df[['UnemploymentRate', 'Employed','LabourParticipationRate']]

# Calculate the correlation matrix
correlation_matrix = heatmap_data.corr()

# Set the figure size
plt.figure(figsize=(4,4))

# Create a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add a title
plt.title('Correlation Heatmap for Unemployment in India')

# Show the plot
plt.show()
```
